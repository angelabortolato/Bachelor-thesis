```python
import numpy as np
import random 
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
import statsmodels.stats.diagnostic
```

TRIAL 0

```python
J0 = np.loadtxt('Jstep0.txt')
```

```python
plt.figure(figsize=(10,8))
plt.title('Matrice di connessione, trial 0', fontsize = 25, pad=25)
plt.xlabel('neuroni')
plt.ylabel('neuroni')
plt.tick_params(top=False, bottom=False, left=False, right=False,
                labelleft=False, labelbottom=False)

plt.matshow(J0,cmap='bwr',fignum=0,aspect='auto')
plt.colorbar()

plt.savefig('matr0.png')
```

```python
plt.rcParams.update({'font.size': 25})
counts, bins = np.histogram(J0,20)
print(counts,bins)
plt.figure(figsize=(11,8))
plt.title('Distribuzione dei pesi, trial 0', pad=25)
plt.xlabel('peso')
plt.ylabel('frequenza')
style = {'facecolor': 'none', 'edgecolor': 'C0', 'linewidth': 2}

plt.hist(bins[:-1], bins, histtype='bar', weights=counts, **style)


bins_center=np.zeros(20)
for i in range((np.shape(bins)[0]-1)):
    bins_center[i]=(bins[i]+bins[i+1])/2


def gauss_function(x, a, x0, sigma):
    return a*np.exp(-(x-x0)**2/(2*sigma**2))
popt, pcov = curve_fit(gauss_function, bins_center, counts, p0 = [1, np.mean(J0), np.std(J0)])
x=np.arange(-0.5, 0.5, 0.01)
plt.plot(x, gauss_function(x, *popt), label='Gaussian fit')
plt.legend(loc=1, frameon=False)


plt.savefig('histo0.png')
```

```python
J0=np.reshape(J0, (40000, 1))

statsmodels.stats.diagnostic.lilliefors(J0, dist='norm', pvalmethod='table')
```

DIFFERENCES

```python
D=J20-J0
```

```python
rel=np.abs(D)/J0

np.mean(rel)
```

```python
plt.figure(figsize=(10,8))
plt.title('Differenze dei pesi pre e post training', fontsize = 25, pad=25)
plt.xlabel('neuroni')
plt.ylabel('neuroni')
plt.tick_params(top=False, bottom=False, left=False, right=False,
                labelleft=False, labelbottom=False)

plt.matshow(D,cmap='bwr',fignum=0,aspect='auto')
plt.colorbar()
plt.savefig('matrdiff.png')
```

```python
plt.rcParams.update({'font.size': 25})
counts, bins = np.histogram(D,20)
print(counts,bins)
plt.figure(figsize=(11,8))
plt.title('Distribuzione delle differenze dei pesi', pad=25)
plt.xlabel(r'$\delta$=differenza di peso')
plt.ylabel('frequenza')
style = {'facecolor': 'none', 'edgecolor': 'C0', 'linewidth': 2}

plt.hist(bins[:-1], bins, histtype='bar', weights=counts, **style)


bins_center=np.zeros(20)
for i in range((np.shape(bins)[0]-1)):
    bins_center[i]=(bins[i]+bins[i+1])/2


def gauss_function(x, a, x0, sigma):
    return a*np.exp(-(x-x0)**2/(2*sigma**2))
popt, pcov = curve_fit(gauss_function, bins_center, counts, p0 = [1, np.mean(D), np.std(D)])
x=np.arange(-0.05, 0.05, 0.001)
plt.plot(x, gauss_function(x, *popt), label='Gaussian fit')
plt.legend(loc=1, frameon=False)


plt.savefig('histodiff.png')
```

```python
D=np.reshape(D, (40000, 1))

statsmodels.stats.diagnostic.lilliefors(D, dist='norm', pvalmethod='table')
```
